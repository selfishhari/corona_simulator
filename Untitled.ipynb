```python
def make_historical_data(province_data, country_data):
    
    sing_data = country_data.loc[(country_data.index== "Singapore"),["Date", "Confirmed"]].sort_values("Date")
    
    uk_data = country_data.loc[(country_data.index== "United Kingdom") \
                    | (country_data.index== "UK"), ["Date", "Confirmed"]].sort_values("Date")
    
    
    aus_data = province_data.loc[province_data.index.str.lower() == "australia", ["Confirmed", "Date"]].sort_values("Date")
    
    aus_data = aus_data.rename({"Confirmed":"Australia"}, axis="columns")
    
    uk_data = uk_data.rename({"Confirmed":"UK"}, axis="columns")
    
    sing_data = sing_data.rename({"Confirmed":"Singapore"}, axis="columns")
    
    hist_data = aus_data.merge(uk_data, on="Date", how="outer").merge(sing_data, on="Date", how="outer")
    
    return hist_data
```

```python
def prepare_data(df):
    
    if "Australia" in df.columns:

        confirmed = ss[['Date','Australia']]        
        
    else:
        
        confirmed = ss[['Date','Confirmed']]
        
    confirmed.columns = ["ds", "y"]

    confirmed["y"] = np.log(confirmed["y"])

    confirmed["y"] = confirmed["y"].replace([np.inf, -np.inf], np.nan).dropna()

    return confirmed


def build_model(historical_data):
    
    m = Prophet(interval_width=0.95, daily_seasonality=True)

    m.fit(historical_data)

    return m

def get_predictions_dataframe(model, horizon=7):
    
    future = model.make_future_dataframe(periods=7)

    future_confirmed = future.copy() # for non-baseline predictions later on

    preds = model.predict(future.tail(horizon))[["yhat", "yhat_lower", "yhat_upper"]]
    
    for x in preds.columns:
        
        preds[x] = np.round(np.exp(preds[x]))

    preds["date"] = future["ds"].tail(horizon).tolist()

    preds.columns = ["confirmed", "lower_bound", "upper_bound", "date"]
    
    history_df = model.history[["ds", "y"]]
    
    history_df["y"] = np.round(np.exp(history_df["y"]))
    
    history_df.columns = ["date", "confirmed"]
    
    history_df["lower_bound"] = history_df["confirmed"]
    
    history_df["upper_bound"] = history_df["confirmed"]
    
    df = pd.concat([history_df, preds], axis=0)

    #preds.loc[:55, "lower_bound"] = np.nan

    #preds.loc[:55, "upper_bound"] = np.nan

    return df


def get_forecasts(historical_data, horizon=7):
    
    prep_hist = prepare_data(historical_data)
    
    model = build_model(prep_hist)
    
    plot_df = get_predictions_dataframe(model, horizon=horizon)
    
    return plot_df
```

```python
def plot_time_series_forecasts(df, country_flag=False):
    
    import plotly


    from plotly.offline import init_notebook_mode, iplot
    import plotly.graph_objs as go
    
    upper_bound = go.Scatter(
    name='Upper Bound',
    x=df['date'],
    y=df['upper_bound'],
    mode='lines',
    marker=dict(color="#444"),
    line=dict(width=0),
    fillcolor='rgba(68, 68, 68, 0.3)',
    fill='tonexty')

    trace = go.Scatter(
        name='Cofirmed Forecast',
        x=df['date'],
        y=df['confirmed'],
        mode='lines',
        line=dict(color='rgb(31, 119, 180)'),
        fillcolor='rgba(68, 68, 68, 0.3)',
        fill='tonexty')

    lower_bound = go.Scatter(
        name='Lower Bound',
        x=df['date'],
        y=df['lower_bound'],
        marker=dict(color="#444"),
        line=dict(width=0),
        mode='lines')
    
    if country_flag:
        

        trace2 = go.Scatter(
            name='UK(confirmed)',
            x=df['date'],
            y=df['UK'],
            mode='lines',
            line=dict(color='rgb(165, 115, 74)'),
            )

        trace3 = go.Scatter(
            name='Singapore(Confirmed)',
            x=df['date'],
            y=df['Singapore'],
            mode='lines',
            line=dict(color='rgb(50, 95, 0)'),
            )
    else:
        
        trace2 = go.Scatter(
            name='Recovered',
            x=df['date'],
            y=df['Recovered'],
            mode='lines',
            line=dict(color='rgb(50, 95, 0)'),
        )

        trace3 = go.Scatter(
            name='Dead',
            x=df['date'],
            y=df['Dead'],
            mode='lines',
            line=dict(color='rgb(165, 115, 74)'),
            )
        

    # Trace order can be important
    # with continuous error bars
    data = [lower_bound, trace, upper_bound, trace2, trace3]

    layout = go.Layout(
        yaxis=dict(title='Case Counts(Cumulative)'),
        title='Historical Trend of Cases with Forecasts',
        showlegend = False)

    fig = go.Figure(data=data, layout=layout)

    iplot(fig, filename= 'shaded_lines')
    
    
    return
    
```
